```python
# Importing Libraries

import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Loading the Data from the csv file to a pandas DataFrame

df=pd.read_csv("/content/steam_games.csv")
df
```

```python
# Selecting the relevant features for recommendation

selected_features=["name","genre","popular_tags","game_details","developer","publisher"]
selected_features
```

```python
# Replacing the null values with null string

for feature in selected_features:
  df[feature]=df[feature].fillna('')
```

```python
df.isna().sum()
```

```python
# Combining all the 5 selected features

combined_features=df["genre"]+' '+df["popular_tags"]+' '+df["game_details"]+' '+df["developer"]+' '+df["publisher"]
combined_features
```

```python
# Converting the text data to feature vectors

vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
feature_vectors
```

```python
# Getting the similarity scores using cosine similarity

similarity=cosine_similarity(feature_vectors)
similarity
```

```python
print(similarity.shape)
```

```python
# Getting the game name from the user

game_name=input("Enter your favourite game name : ")
```

```python
# Creating a list with all the game names given in the dataset

list_of_all_names=df['name'].tolist()
print(list_of_all_names)
```

```python
# Finding the close match of the game name given by the user

find_close_match=difflib.get_close_matches(game_name,list_of_all_names)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
# Finding the index of the game with title

index_of_the_game=df[df.name==close_match]['index'].values[0]
print(index_of_the_game)
```

```python
# Getting a list of similar games

similarity_score=list(enumerate(similarity[index_of_the_game]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# Sorting the games based on their similarity score

sorted_similar_games=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_games)
```

```python
# Print the name of similar games based on the index

print('Games suggested for you : \n')
i=1
for game in sorted_similar_games:
  index=game[0]
  name_from_index=df[df.index==index]['name'].values[0]
  if (i<30):
    print(i,'.',name_from_index)
    i+=1
```
